```python
from sklearn.utils import shuffle
Xtrain_shuffled, ytrain_shuffled = shuffle(Xtrain,ytrain, random_state = 9896)
Xtest_shuffled, ytest_shuffled = shuffle(Xtest,ytest, random_state = 9896)
```

```python
Xtrain_normalized = Xtrain/255
Xtest_normalized = Xtest/255
Xvalid_normalized = Xvalid/255
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(ytrain,10)
y_test = to_categorical(ytest,10)
y_valid = to_categorical(yvalid,10)
```

```python
from keras import layers , models
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
tf.random.set_seed(2)

model = models.Sequential()
model.add(layers.Conv2D(32,3,activation = 'relu',input_shape = (150,150,3)))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(64,3,activation = 'relu'))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(128,3,activation = 'relu'))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(256,3,activation = 'relu'))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(512,3,activation = 'relu'))
model.add(layers.MaxPooling2D())

model.add(layers.Dropout(0.25))
model.add(layers.Flatten())
model.add(layers.Dense(512,activation = 'relu'))
model.add(layers.Dense(10,activation = 'softmax'))

# model = models.Sequential()
# model.add(layers.Conv2D(128,3,activation = 'relu',input_shape = (150,150,3)))
# model.add(layers.MaxPooling2D(3,3))
# model.add(layers.Conv2D(512,3,activation = 'relu'))
# model.add(layers.MaxPooling2D(3,3))
# model.add(layers.Conv2D(1028,3,activation = 'relu'))

# # model.add(layers.Conv2D(510,3,activation = 'relu'))
# # model.add(layers.Conv2D(510,3,activation = 'relu'))
# model.add(layers.MaxPooling2D(3,3))
# model.add(layers.Dropout(0.4))

# model.add(layers.Flatten())
# model.add(layers.Dense(512,activation = 'relu'))
# model.add(layers.Dense(20,activation = 'softmax'))
```

```python
model.compile(optimizer=Adam(0.00001),loss = 'categorical_crossentropy',metrics = ['acc'])
```

```python
hist = model.fit(Xtrain_normalized,y_train ,epochs=30,validation_data=(Xvalid_normalized,y_valid))
```

```python
model.evaluate(Xtest,y_test)
```

```python
histo = hist.history
histo.keys()
```

```python
epochs = np.arange(1,len(histo['val_acc'])+1)
import matplotlib.pyplot as plt
plt.title('model accuracy')
plt.xlabel("Epochs")
plt.ylabel("Training and Validation Accuracy")
plt.plot(epochs,histo["acc"])
plt.plot(epochs,histo["val_acc"])
plt.legend(['train', 'val'], loc='upper left')
```

```python
epochs = np.arange(1,len(histo['val_acc'])+1)
import matplotlib.pyplot as plt
plt.xlabel("Epochs")
plt.ylabel("Training and Validation Loss")
plt.plot(epochs,histo["loss"])
plt.plot(epochs,histo["val_loss"])
plt.title('model loss')
plt.legend(['train', 'val'], loc='upper left')
```
